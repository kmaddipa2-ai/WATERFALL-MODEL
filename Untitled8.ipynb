```python
"""
Weather Modeling using Quadratic Equation
Following the Waterfall Model phases
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def quadratic_weather_model(time: np.ndarray, a: float, b: float, c: float) -> np.ndarray:
    """Quadratic model: f(t) = a*t^2 + b*t + c"""
    return a * (time ** 2) + b * time + c

def load_coefficients(file_path: str) -> pd.DataFrame:
    """Load coefficients (a, b, c) from CSV"""
    df = pd.read_csv(file_path)
    required_cols = {"a", "b", "c"}
    if not required_cols.issubset(df.columns):
        raise ValueError(f"CSV must contain columns: {required_cols}")
    return df


def test_quadratic_weather_model():
    """Simple test for quadratic function"""
    t = np.array([0, 1, 2])
    result = quadratic_weather_model(t, 1, 2, 1)
    expected = np.array([1, 4, 9])
    assert np.allclose(result, expected), "Test failed!"
    print("Unit Test Passed!")

def simulate_and_plot(df: pd.DataFrame, days: int):
    """Simulate temperature and display plots in Colab"""

    time = np.arange(0, days + 1, 1)

    for idx, row in df.iterrows():
        a, b, c = row["a"], row["b"], row["c"]
        temps = quadratic_weather_model(time, a, b, c)

        plt.figure()
        plt.plot(time, temps, marker="o", label=f"Model {idx+1} (a={a}, b={b}, c={c})")
        plt.xlabel("Time (days)")
        plt.ylabel("Temperature")
        plt.title("Weather Modeling - Quadratic Variation")
        plt.legend()
        plt.grid(True)
        plt.show()

def main():
    print("=== Weather Modeling (Waterfall Method) ===")

    file_path = "abc_values.csv"
    days = 30

    df = load_coefficients(file_path)

    test_quadratic_weather_model()

    simulate_and_plot(df, days)

if __name__ == "__main__":
    main()
```
